# Análise Demográfica dos entregadores que saíram da plataforma (Churns)

```python

# Análise demográfica por genêro (Masculino ou Feminino)
df_churned_gender_count = df_churned.groupby('GENERO').count().reset_index()
df_churned_gender_count.loc[df_churned_gender_count.GENERO=='M','GENERO'] = 'Masculino'
df_churned_gender_count.loc[df_churned_gender_count.GENERO=='F','GENERO'] = 'Feminino'
df_churned_gender_count.loc[df_churned_gender_count.GENERO=='O','GENERO'] = 'Outro'

# fig = px.bar(df_churned_gender_count, y='ID', title="Gênero dos entregadores que deram 'Churn'")
fig = px.pie(df_churned_gender_count, values='ID', names='GENERO', title="Gênero dos entregadores que deram 'Churn'")
fig.show()
```

# Análise Demográfica dos que entregadores que permaneceram na plataforma

```python
# Análise demográfica por genêro (Masculino ou Feminino)
df_not_churned_gender_count = df_not_churned.groupby('GENERO').count().reset_index()
df_not_churned_gender_count.loc[df_not_churned_gender_count.GENERO=='M','GENERO'] = 'Masculino'
df_not_churned_gender_count.loc[df_not_churned_gender_count.GENERO=='F','GENERO'] = 'Feminino'
df_not_churned_gender_count.loc[df_not_churned_gender_count.GENERO=='O','GENERO'] = 'Outro'

#fig = px.bar(df_not_churned_gender_count, y='ID', x='GENERO', title="Gênero dos entregadores que deram 'Churn'")
fig = px.pie(df_not_churned_gender_count, values='ID', names='GENERO', title="Gênero dos entregadores que permaneceram na plataforma")
fig.show()
```

# Comparação Churn vs. Não Churn

```python
df_churned.loc[:, 'TOTAL_ORDERS'] = df_churned['COUNT_ORDERS_RESTAURANTES'] + df_churned['COUNT_ORDERS_MERCADO'] + df_churned[ 'COUNT_ORDERS_FARMACIA'] + df_churned['COUNT_ORDERS_EXPRESS'] + df_churned['COUNT_ORDERS_ECOMMERCE'] + df_churned['COUNT_ORDERS_ANTOJO']
df_not_churned.loc[:, 'TOTAL_ORDERS'] = df_not_churned['COUNT_ORDERS_RESTAURANTES'] + df_not_churned['COUNT_ORDERS_MERCADO'] + df_not_churned[ 'COUNT_ORDERS_FARMACIA'] + df_not_churned['COUNT_ORDERS_EXPRESS'] + df_not_churned['COUNT_ORDERS_ECOMMERCE'] + df_not_churned['COUNT_ORDERS_ANTOJO']

df_churned_stats = df_churned.describe()
df_not_churned_stats = df_not_churned.describe()

df_churned_comparison = df_churned_stats.compare(df_not_churned_stats)[['TOTAL_ORDERS', 'GORJETA', 'FRETE_MEDIO', 'COOKING_TIME_MEDIO', 'ITENS_MEDIO']]
df_churned_comparison = df_churned_comparison.T.reset_index()

df_churned_comparison.loc[df_churned_comparison['level_1'] =='self','level_1'] = 'CHURN'
df_churned_comparison.loc[df_churned_comparison['level_1'] =='other','level_1'] = 'PERMANECE'
df_churned_comparison['X'] = df_churned_comparison['level_0'] + df_churned_comparison['level_1']
df_churned_comparison_orders = df_churned_comparison[df_churned_comparison["level_0"] == 'TOTAL_ORDERS']
df_churned_comparison_tips = df_churned_comparison[df_churned_comparison["level_0"] == 'GORJETA']
df_churned_comparison_freight = df_churned_comparison[df_churned_comparison["level_0"] == 'FRETE_MEDIO']
df_churned_comparison_cooking_time = df_churned_comparison[df_churned_comparison["level_0"] == 'COOKING_TIME_MEDIO']
df_churned_comparison_items = df_churned_comparison[df_churned_comparison["level_0"] == 'ITENS_MEDIO']

df_churned_comparison
tags = {
  "level_1": "Status do Entregador",
  "mean": "Média",
}
fig = px.bar(df_churned_comparison_orders, x='level_1', y='mean', title="Número de Pedidos Realizados (Média)", labels=tags)
fig.show()
fig = px.bar(df_churned_comparison_tips, x='level_1', y='mean', title="Total de Gorjetas Recebidas (Média)", labels=tags)
fig.show()
fig = px.bar(df_churned_comparison_freight, x='level_1', y='mean', title="Tempo dos Pedidos (Média)", labels=tags)
fig.show()
fig = px.bar(df_churned_comparison_cooking_time, x='level_1', y='mean', title="Tempo de Espera no Restaurante (Média)", labels=tags)
fig.show()
fig = px.bar(df_churned_comparison_items, x='level_1', y='mean', title="Itens Entregues (Média)", labels=tags)
fig.show()
```
